#### Test Train Split

```python
from sklearn.model_selection import train_test_split

# Separate features and target
features = data.drop('NObeyesdad', axis=1)
target = data['NObeyesdad']

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=42)

# Print the shapes of the train and test sets
print('Shape of X_train:', X_train.shape)
print('Shape of X_test:', X_test.shape)
print('Shape of y_train:', y_train.shape)
print('Shape of y_test:', y_test.shape)
```

#### Define Models

```python
# import logistic regression, gaussianNB, Tree and set into model1, model2, model 3

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

model1 = LogisticRegression()
model2 = GaussianNB()
model3 = DecisionTreeClassifier()
```

#### Model Tuning

```python
#Tune model 1 and 3 with gridsearch cv
from sklearn.model_selection import GridSearchCV

# Define the grid of hyperparameters to search for model1
param_grid1 = {
    'penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'max_iter': [100, 200, 300]
}

# Define the grid of hyperparameters to search for model3
param_grid3 = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialize GridSearchCV for model1
grid_search1 = GridSearchCV(model1, param_grid1, cv=5)

# Initialize GridSearchCV for model3
grid_search3 = GridSearchCV(model3, param_grid3, cv=5)

# Fit the grid search objects to the data
grid_search1.fit(X_train, y_train)
grid_search3.fit(X_train, y_train)

# Get the best models
best_model1 = grid_search1.best_estimator_
best_model3 = grid_search3.best_estimator_

# Print the best parameters for each model
print("Best parameters for model1:", grid_search1.best_params_)
print("Best parameters for model3:", grid_search3.best_params_)
```

#### Voting Classifier

```python
# import voting classifier
from sklearn.ensemble import VotingClassifier

# Create a voting classifier
model4 = VotingClassifier(estimators=[('LogisticRegression', best_model1),
                                    ('GaussianNB', model2),
                                    ('DecisionTreeClassifier', best_model3)],
                         voting='hard')
```

```python
#Fits models on the training data.

best_model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
best_model3.fit(X_train, y_train)
model4.fit(X_train, y_train)
```

### Prediction

```python
# prediction

from sklearn import metrics
y_pred1 = best_model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred3 = best_model3.predict(X_test)
y_pred4 = model4.predict(X_test)
print("Logistic Regression Accuracy:", metrics.accuracy_score(y_test, y_pred1))
print("GaussianNB Accuracy:", metrics.accuracy_score(y_test, y_pred2))
print("DecisionTreeClassifier Accuracy:", metrics.accuracy_score(y_test, y_pred3))
print("VotingClassifier Accuracy:", metrics.accuracy_score(y_test, y_pred4))
```
